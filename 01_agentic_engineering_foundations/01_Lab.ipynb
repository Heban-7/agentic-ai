# Agentic AI Engineering Foundation 

## Calling LLM through API keys

```python
# Let's start by importing the necessary libraries
from dotenv import load_dotenv
import os
import openai
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
# load the environment variables
load_dotenv(override=True)

# set the API key
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
#create instance of openai
openai = OpenAI()

# create a function for making a request to the OpenAI API
def make_request(messages):
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages
    )
    return response.choices[0].message.content
```

```python
# create a message for the OpenAI API
user_input = "Hello, Welcome to the course"
messages = [{'role': 'user', 'content': user_input}]

# make a request to the OpenAI API
response = make_request(messages)
print(response)
```

```python
# Now ask for a question:
question = "Please propose a hard, challenging question to assess someone's IQ. Respond only with the question."
messages = [{"role": "user", "content": question}]
```

```python
# make a request to the OpenAI API
response = make_request(messages)
print(response)
```

```python
# let's ask the LLM to answer the question
messages = [{"role": "user", "content": response}]
response = make_request(messages)
print(response)
```

```python
# let's make the response in md format
from IPython.display import Markdown, display
display(Markdown(response))
```

```python
# let's ask another question in quantum mechanics
question = "what is superposition and entanglement in quantum mechanics?"
messages = [{'role': 'user', 'content': question}]

#make a request to the OpenAI API
response = make_request(messages)

# print the response in md format
display(Markdown(response))
```

```python
# Lets make a business idea: first ask the llm to pick a business area that might be worth exploring for an Agentic AI business AI opportunity
question1 = "Please propose a business idea for an Agentic AI business. Respond only with the business idea."
messages1 = [{"role": "user", "content": question1}]

# make a request to the OpenAI API
business_idea = make_request(messages1)

# print the business idea in md format
display(Markdown(business_idea))
```

```python
# Now let's ask the LLM to present a pain point in that industry- something challenging that might ve ripe for an Agentic solution
question2 = f"Please present a pain point in in this Agentic AI business area. here is my business idea: {business_idea}. Respond only with the pain point."
messages2 = [{"role": "user", "content": question2}]

# make a request to the OpenAI API
pain_point = make_request(messages2)

# print the pain point in md format
display(Markdown(pain_point))
```

```python
# Now let's ask the LLM to propose a solution to the pain point
question3 = f"Please propose a solution to the pain point: {pain_point}. for the business idea: {business_idea}. Respond only with the solution."
messages3 = [{"role": "user", "content": question3}]

# make a request to the OpenAI API
solution = make_request(messages3)

# print the solution in md format
display(Markdown(solution))
```
