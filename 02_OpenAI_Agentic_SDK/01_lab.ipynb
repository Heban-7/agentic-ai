**OpenAI Agents SDK:** lightweight, production-ready Agentic AI framework

```python
# import basic libraries
from dotenv import load_dotenv
from agents import Agent, Runner, trace

# load the environment variables
load_dotenv(override=True)
```

```python
agent = Agent(name="Joke Teller", instructions="You're a joke teller. You're given a topic and you're going to tell a joke about it.", model="gpt-4o-mini")
message = "Tell me a joke about quantum intagallement, don't make your joke like question and answer, make it like a story"
runner = await Runner.run(agent, message)

print(runner.final_output)
```

```python
# Make an agent and run it and check the trace
agent = Agent(name="Joke Teller", instructions="You're a joke teller. You're given a topic and you're going to tell a joke about it.", model="gpt-4o-mini")
message = "Tell me a joke about quantum intagallement, don't make your joke like question and answer, make it like a story"

with trace("Telling a quantum intagallement joke"):
    result = await Runner.run(agent, message)
    print(result.final_output)
```
